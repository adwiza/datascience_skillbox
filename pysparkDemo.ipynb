```python
# path_to_my_file = "/Users/adwiz/Downloads/_c4s_events_2023_06_01__202306201703.csv.gz"
path_to_my_file = 'datasets/all_events.csv'

df = (spark.read      
    .option("header", "true")
    .option("delimiter", ",")   
    .option('nullValue', None)
    .schema(schema)
    .format("csv")
    .option("compression", "gzip")
    .load(path_to_my_file)
#   .csv(path_to_my_file, inferSchema='true') #.limit(10000000)
)

df.printSchema()
```

```python
df.select(col("`customer.device.family`")).show()
```

```python
df.select(df.colRegex("`^.*studio.name*`")).show(truncate=False)
# df.select(col("`clip.studio.name`"),col("`studio.name`")).show(1000, truncate=False)
```
