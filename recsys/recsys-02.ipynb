```python
def Precision_at_n(df_ratings, n=3, threshold=5):
    rel_n = set(df_ratings[df_ratings.real_rating>=threshold]['item'])
    t=df_ratings.sort_values(by='rating', ascending=False)[0:n]
    ret_n=set(t['item'])
    return len(rel_n & ret_n)/n
```

```python
def Avg_Precision_at_n(df_ratings, n=10,threshold=5):
    avg_p=0
    rel_n = set(df_ratings[df_ratings.rating>=threshold]['item'])
    
    for i in range(1,n+1):
        t=df_ratings.sort_values(by='rating', ascending=False)[0:i]
        ret_n=set(t['item'])
        t2=df_ratings.sort_values(by='rating', ascending=False)[i-1:i]
        ret_n2=set(t2['item'])
        #print(ret_n2,ret_n)
        if len((ret_n2 & rel_n))>0: #rel(i)
            avg_p=avg_p+len(rel_n & ret_n)/n
    return avg_p/n
```

```python
 def MAP_at_n(df_ratings_all, n=10, threshold=5):
        users = df_ratings_all['user'].unique()
        map_at_n = 0
        for i in users:
            df_ratings = df_ratings_all[df_ratings_all.user == i]
            map_at_an = map_at_n+Avg_Precision_at_n(df_ratings, n=n, threshold=threshold)
        return map_at_n/len(users)
```
