# Requests

```python
import requests
import pandas as pd
```

# GET

```python
url = 'https://itunes.apple.com/search'
params = {'media': 'music', 'term': 'Manizha', 'limit': 30}
response = requests.get(url, params=params, timeout=10)
```

```python
response.status_code
```

```python
response.json()
```

```python
results = response.json()['results']
results
```

```python
df = pd.DataFrame(results)
df
```

```python
df.primaryGenreName.value_counts()
```

```python
help(df.to_excel)
```

```python
df.to_excel("datasets/iTunes_api.xlsx", sheet_name="Manizha", index_label='id')
```

```python
help(df.to_csv)
```

```python
df.to_csv("datasets/iTunes_api.csv", sep='\t', index=False)
```

# POST

```python
url = 'https://itunes.apple.com/search'
params = {'media': 'music', 'term': 'Manizha', 'limit': 30}
response = requests.post(url, data=params, timeout=10)
response.status_code
```
