```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_excel('datasets/Когортный анализ.xlsx')
```

```python
data.head()
```

```python
def get_month(x):
    return dt.datetime(x.year, x.month, 1)
```

```python
data['month_year'] = data['date'].apply(get_month)
```

```python
data.head()
```

```python
data['cohort_month'] = data.groupby('user_id')['month_year'].transform('min')
```

```python
data.head(50)
```

```python
def get_date(df, column):
    year = df[column].dt.year
    month = df[column].dt.month
    day = df[column].dt.day
    return year, month, day
```

```python
date_year, date_month, _ = get_date(data, 'date')
```

```python
cohort_year, cohort_month, _ = get_date(data, 'cohort_month')
```

```python
year_diff = date_year - cohort_year
```

```python
month_diff = date_month - cohort_month
```

```python
data['cohort_index'] = year_diff * 12 + month_diff + 1
```

```python
data.head()
```

```python
data['cohort_index'].unique()
```

```python
pivot_table = data.pivot_table(index='cohort_month',
                               columns='cohort_index',
                               values='user_id',
                               aggfunc=len)
```

```python
pivot_table.head(20)
```

```python
first_column = pivot_table.iloc[:, 0]
```

```python
retention = pivot_table.div(first_column, axis=0)
```

```python
retention.round(2)
```

```python
plt.figure(figsize=(11, 9))
plt.title('Когортный анализ Retention')
sns.heatmap(data=retention, 
            annot=True,
            fmt='.0%',
            vmin=.0,
            vmax=.5,
            cmap='YlGnBu'
           );
```
