```python
import json
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# json_file = 'datasets/test_data.json'
json_file = 'datasets/dataset.json'
with open(json_file,'r') as f:
    data = json.load(f)
```

```python
df = pd.DataFrame(data, columns=['text', 'hasBadWords'])
df.hasBadWords = df.hasBadWords.apply(lambda x: str(x) if x == True else str(0))
# df['text'] = df['text'].apply(lambda x: str(x).split(' '))
```

```python
contain_values = df[df['hasBadWords'].str.contains('0')]
print (contain_values)
```

```python
contain_values.to_csv('datasets/check_false_sentenses.csv')
```

```python
count_vectorizer = CountVectorizer()
lowercase_list = [str(e) for e in df['text']]
bag_of_words = count_vectorizer.fit_transform(lowercase_list)
bag_of_words = pd.DataFrame(bag_of_words.toarray(),
columns = count_vectorizer.get_feature_names_out())
```
