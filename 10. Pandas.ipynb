```python
pdnp_df = pd.read_csv('datasets/partner_data_next_part.csv')
pdnp_df.head()
```

```python
pdnp_df.shape
```

```python
full_df = pd.concat([pd_df, pdnp_df])
```

```python
full_df.tail(15)
```

```python
full_df = pd.concat([pd_df, pdnp_df], axis=1)
full_df.tail(15)
```
